```python
import pandas as pd
import numpy as np
```

```python
# Преобразуем в даты и используем NumPy
forum_users = {
    'User ID': np.array([1,2,3,4,5]),
    'Username': ['bogdan_s','jane_smith','alex123','bob56','mark_wilson'],
    'Age': np.array([18,35,25,38,None]),
    'Joined Date': pd.to_datetime(['2023-01-01','2023-02-01','2023-03-01','2023-04-01','2023-05-01']),
    'Total Posts': np.array([150,230,80,420,310]),
    'Reputation': np.array([500,720,200,940,500])
}
df = pd.DataFrame(forum_users)
df
```

## Add / Edit

```python
df['Languages'] = 'English' # Add all
df
```

```python
df['Reputation'] = 100 # edit all
df
```

```python
df['Reputation'] = [100,202,303,404,505] # edit all
df
```

```python
df['Active'] = [True,False,False, True, True]
df
```

## Merge

```python
total_comments = {
    'Total Comments': [70,30,45,55,80]
}
# df2 = pd.DataFrame(total_comments,index=['a','b','c','d','e'])
df2 = pd.DataFrame(total_comments)
df2
```

```python
pd.concat([df,df2]) # Контатынация или слияния по оси 0
```

```python
pd.concat([df,df2],axis=1) # Контатынация или слияния по оси 1
```
