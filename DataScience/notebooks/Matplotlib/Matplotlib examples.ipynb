```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
values = [1,2,3,4,5,6,7,8,9]
squares = [ value ** 2 for value in values]
plt.plot(values,squares)
plt.show()
```

```python
plt.scatter(values,squares)
plt.show()
```

```python
plt.bar(values,squares)
plt.show()
```

```python
plt.stem(values,squares)
plt.show()
```

```python
values2 = list(range(100))
pows = [pow(value,3) for value in values2]
plt.scatter(values2,pows)
plt.show()
```

```python
plt.scatter(values2,pows, c=pows, cmap=plt.cm.Blues, edgecolors='none',s=10)
plt.show()
```

```python
values = list(range(50))
squares = [pow(value,2) for value in values]
cubes = [pow(value,3) for value in values]
fig, (ax1,ax2) = plt.subplots(2,1,sharex=True) # 2 по вертикали, 1 по горизонтали, sharex=True = х будет общий
ax1.scatter(values,squares)
ax1.set_title('Square numbers')
ax2.plot(values,cubes)
ax2.set_title('Cubes numbers')
plt.show()
```

```python
values = list(range(10))
squares = [pow(value,2) for value in values]
cubes = [pow(value,3) for value in values]
fig, (ax1,ax2) = plt.subplots(1,2,sharey=True) # Изменили параметры
ax1.scatter(values,squares)
ax1.set_title('Square numbers')
ax2.plot(values,cubes)
ax2.set_title('Cubes numbers')
plt.show()
```

```python
data = {
    'Year':[2001,2002,2003,2004,2005,2006],
    'Sales':[100,150,130,180,230,270]
}
df  = pd.DataFrame(data)
df.plot(x='Year',y='Sales',kind='line')
plt.title('Year Sales')
plt.show()
```

```python
data = {
    'Category': ['FIRST','FIRST','SECOND','SECOND','THIRD','THIRD','SECOND'],
    'Value':[10,20,25,45,45,50,50]
}
df = pd.DataFrame(data)
df
```

```python
# Диапозоный график
df.boxplot(by='Category',column='Value')
plt.title('Values distribution by category')
plt.show()
```

```python
data = {
    'Month':['Jan','Feb','Mar','Apr','May'],
    'Week 1 Temp':[6,10,12,14,17],
    'Week 2 Temp':[8,12,13,14,18],
    'Week 3 Temp':[9,11,14,16,20],
    'Week 4 Temp':[10,15,10,17,22],
}
df = pd.DataFrame(data)
df.head()
```

```python
df.set_index('Month',inplace=True) # index по оси 0 стал Month
df
```

```python
sns.heatmap(df, cmap='coolwarm',annot=True)
plt.title('Weekly temperature data')
plt.show()
```
